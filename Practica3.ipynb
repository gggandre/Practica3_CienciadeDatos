```python
#Escriba un programa que pregunte al usuario por el número de horas trabajadas y el costo por hora. Mostrar en pantalla la paga que resulta
horastrabajadas = float(input("Introduce el número de horas trabajadas: "))
costoporhora = float(input("Introducir el costo por hora: "))
pagaresultante = horastrabajadas * costoporhora
print(pagaresultante)
```

```python
#Escriba un programa que pida al usuario su peso en kg, y su estatura en m, calcule el indice de masa corporal
#, lo almacene en una variable e imprima la frase "Tu indice de masa corporal"
import math
 
peso = float(input("Ingrese su peso en Kilogramos: "))
estatura = float(input("Ingrese su estatura en metros: "))
 
IMC = round(peso/math.pow(estatura,2),1)
 
print("Tu indice de masa corporal es de: "+str(IMC))
```
